```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset=pd.read_csv('/Users/Aayushi/Desktop/ML/Regression/Multiple_Linear_Regression/50_Startups.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,4].values
dataset

#categorical data is found in the second last column
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_x=LabelEncoder()
x[:,3]=labelencoder_x.fit_transform(x[:,3])
onehotencoder=OneHotEncoder(categorical_features=[3])
x=onehotencoder.fit_transform(x).toarray()
#convert scientific notation in output
x.astype(int)

#avaoiding the dummy variable trap
x=x[:,1:]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)

#fitting multiple reg to the training set
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)

#prediction of test set results
y_pred = regressor.predict(x_test)
y_pred,y_test

#building the optional model using backward elimination
#we add a columns of ones so that the sm library doesn't
#think of our eq is without a constant b0
#x = np.append(arr = x, values= np.ones((50,1)).astype(int),axis=1)
#this coloumn will be inserted at last so we instead add the array x to this coloumn of ones
import statsmodels.formula.api as sm
x = np.append(arr = np.ones((50,1)).astype(int), values= x,axis=1)

x_opt=x[:,[0,1,2,3,4,5]]
x_opt.astype(int)
regressor_ols = sm.OLS(endog=y ,exog=x_opt).fit()
regressor_ols.summary( )
#highest p value corresponds to x2 ,so we remove it
x_opt=x[:,[0,1,3,4,5]]
x_opt.astype(int)
regressor_ols = sm.OLS(endog=y ,exog=x_opt).fit()
regressor_ols.summary( )

x_opt=x[:,[0,3,4,5]]
x_opt.astype(int)
regressor_ols = sm.OLS(endog=y ,exog=x_opt).fit()
regressor_ols.summary( )

x_opt=x[:,[0,3,5]]
x_opt.astype(int)
regressor_ols = sm.OLS(endog=y ,exog=x_opt).fit()
regressor_ols.summary( )

x_opt=x[:,[0,3]]
x_opt.astype(int)
regressor_ols = sm.OLS(endog=y ,exog=x_opt).fit()
regressor_ols.summary( )
```
